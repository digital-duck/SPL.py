# Eigenvalues And Eigenvectors

Eigenvalues and eigenvectors provide a fundamental way to understand the behavior of linear transformations represented by matrices. Specifically, an eigenvector of a matrix remains unchanged in direction when the transformation is applied, merely scaled by a factor known as its corresponding eigenvalue. These concepts are crucial for analyzing stability, diagonalization, and various other applications in linear algebra. The worked example below uses the matrix A = [[4, 1], [2, 3]].

````python
import sympy as sp

A = sp.Matrix([[4, 1], [2, 3]])
print('Matrix A:')
A
````

## Eigendecomposition

Compute eigenvalues and eigenvectors exactly using SymPy:

````python
eigendata = A.eigenvects()
for lam, mult, vecs in eigendata:
    for v in vecs:
        print(f'λ = {lam},  v = {v.T}')
````

**Computed by the SPL kernel during workflow execution:**

```
lambda=2 → v=[-1/2, 1]
lambda=5 → v=[1, 1]
```

## What Do These Values Mean?

The eigenvalues of A (λ = 2 and λ = 5) represent the scaling factors along the corresponding eigenvectors when the matrix transforms vectors. Specifically, the eigenvector v=[-1/2, 1] is a direction that is scaled by a factor of 2 when transformed by A, while the eigenvector v=[1, 1] is scaled by a factor of 5 in that same direction.  In essence, these eigenvalues and eigenvectors define how the linear transformation represented by matrix A stretches or compresses vectors in different directions.

## Verification: A·v = λ·v

Confirm each eigenpair satisfies the defining equation:

````python
for lam, mult, vecs in eigendata:
    for v in vecs:
        result = A * v == lam * v
        print(f'λ={lam}: A·v == λ·v → {result}')
````
