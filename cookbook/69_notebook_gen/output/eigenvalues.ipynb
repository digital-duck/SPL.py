# Eigenvalues And Eigenvectors

Eigenvalues and eigenvectors are special scalar-vector pairs associated with a square matrix that reveal its most fundamental geometric behavior: when a matrix acts on certain vectors, the direction of those vectors is preserved and only their magnitude is scaled. Formally, for a square matrix **A**, an eigenvector **v** is a nonzero vector satisfying **Av** = λ**v**, where the scalar λ is the corresponding eigenvalue — intuitively, the matrix simply "stretches" or "flips" that vector by a factor of λ. Finding these pairs is central to applications ranging from principal component analysis to solving systems of differential equations, and the process begins by solving the characteristic equation det(**A** − λ**I**) = 0. The worked example below uses the matrix A = [[4, 1], [2, 3]].

````python
import sympy as sp

A = sp.Matrix([[4, 1], [2, 3]])
print('Matrix A:')
A
````

## Eigendecomposition

Compute eigenvalues and eigenvectors exactly using SymPy:

````python
eigendata = A.eigenvects()
for lam, mult, vecs in eigendata:
    for v in vecs:
        print(f'λ = {lam},  v = {v.T}')
````

```
lambda=2 → v=[-1/2, 1]
lambda=5 → v=[1, 1]
```

## What Do These Values Mean?

The eigenvectors **[-1/2, 1]** and **[1, 1]** identify the two special directions in the plane that the transformation A leaves invariant — vectors along these directions are only stretched or compressed, never rotated or sheared. Under A, a vector pointing along **[1, 1]** is scaled by **λ=5**, meaning it gets stretched to five times its original length, while a vector along **[-1/2, 1]** is scaled by **λ=2**, doubling in length. Together, these eigenpairs reveal the "skeleton" of the transformation: any vector in the plane can be decomposed into these two directions, and the transformation simply multiplies each component independently by its corresponding eigenvalue.

## Verification: A·v = λ·v

Confirm each eigenpair satisfies the defining equation:

````python
for lam, mult, vecs in eigendata:
    for v in vecs:
        result = A * v == lam * v
        print(f'λ={lam}: A·v == λ·v → {result}')
````
