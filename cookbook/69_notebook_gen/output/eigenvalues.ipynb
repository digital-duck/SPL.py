# Eigenvalues And Eigenvectors

In linear algebra, eigenvalues and eigenvectors are fundamental concepts that help describe the behavior of matrices. An eigenvector is a non-zero vector that, when a matrix is multiplied by it, results in a scaled version of itself. The scalar value that multiplies this result is known as the eigenvalue, which can provide valuable information about the stability and properties of a system.

The worked example below uses the matrix A = [[4, 1], [2, 3]].

````python
import sympy as sp

A = sp.Matrix([[4, 1], [2, 3]])
print('Matrix A:')
A
````

## Eigendecomposition

Compute eigenvalues and eigenvectors exactly using SymPy:

````python
eigendata = A.eigenvects()
for lam, mult, vecs in eigendata:
    for v in vecs:
        print(f'λ = {lam},  v = {v.T}')
````


## What Do These Values Mean?

The eigenvector $v = [-1/2, 1]$ represents a direction that remains unchanged when transformed by matrix A. The corresponding eigenvalue $\lambda = 2$ indicates that this direction is scaled by a factor of 2. This means that if you were to apply the transformation defined by matrix A to an object along this direction, it would be stretched or compressed by a factor of 2.

The eigenvector $v = [1, 1]$ represents another direction that remains unchanged when transformed by matrix A. The corresponding eigenvalue $\lambda = 5$ indicates that this direction is scaled by a factor of 5. This means that if you were to apply the transformation defined by matrix A to an object along this direction, it would be stretched or compressed by a factor of 5.

The eigenvector $v = [-1/2, 1]$ and the eigenvector $v = [1, 1]$ are orthogonal, which means that they represent directions on which the transformation leaves objects unchanged with respect to each other.

## Verification: A·v = λ·v

Confirm each eigenpair satisfies the defining equation:

````python
for lam, mult, vecs in eigendata:
    for v in vecs:
        result = A * v == lam * v
        print(f'λ={lam}: A·v == λ·v → {result}')
````
