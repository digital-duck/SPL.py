# Dot Products And Orthogonality

The dot product of two vectors measures how much they point in the same direction, computed by multiplying corresponding components and summing the results — for vectors **u** and **v**, this gives a single scalar value u·v. When two vectors have a dot product of zero, they are said to be *orthogonal*, meaning they meet at a right angle, which is a property that appears throughout linear algebra in decompositions, projections, and transformations. The worked example below uses the matrix A = [[1, 0], [0, -1]].

````python
import sympy as sp

A = sp.Matrix([[1, 0], [0, -1]])
print('Matrix A:')
A
````

## Eigendecomposition

Compute eigenvalues and eigenvectors exactly using SymPy:

````python
eigendata = A.eigenvects()
for lam, mult, vecs in eigendata:
    for v in vecs:
        print(f'λ = {lam},  v = {v.T}')
````

```
lambda=-1 → v=[0, 1]
lambda=1 → v=[1, 0]
```

## What Do These Values Mean?

The eigenvectors **[1, 0]** and **[0, 1]** are the x- and y-axes respectively — the two directions in the plane that this transformation leaves unchanged in orientation. Under the matrix A, any vector along the x-axis is scaled by **+1** (it stays exactly where it is), while any vector along the y-axis is scaled by **-1** (it is reflected through the origin, flipping to point in the opposite direction). Geometrically, A is a reflection across the x-axis: horizontal vectors are fixed, and vertical vectors are inverted.

## Verification: A·v = λ·v

Confirm each eigenpair satisfies the defining equation:

````python
for lam, mult, vecs in eigendata:
    for v in vecs:
        result = A * v == lam * v
        print(f'λ={lam}: A·v == λ·v → {result}')
````
